##### Binary Step Activation Function

```python
def binaryStep(x):
    ''' It returns '0' is the input is less then zero otherwise it returns one '''
    return np.heaviside(x,1)
```

```python
x = np.linspace(-10, 10)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.plot(x, binaryStep(x))
plt.axis('tight')
plt.title('Activation Function : BinaryStep')
plt.show()
```

##### Linear Activation Function

```python
def linear(x):
    ''' y = f(x) It returns the input as it is'''
    return x
```

```python
x = np.linspace(-10, 10)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.plot(x, linear(x))
plt.axis('tight')
plt.title('Activation Function : Linear')
plt.show()
```

##### Sigmoid Activation Function

```python
def sigmoid(x):
    ''' It returns 1/(1+exp(-x)). where the values lies between zero and one '''

    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10, 10)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function : Sigmoid')
plt.show()
```

##### Tanh Activation Function

```python
def tanh(x):
    ''' It returns the value (1-exp(-2x))/(1+exp(-2x)) and the value returned will be lies in between -1 to 1.'''

    return np.tanh(x)
```

```python
x = np.linspace(-10, 10)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function : Tanh')
plt.show()
```

##### ReLu Activation Function

```python
def ReLu(x):
    ''' It returns zero if the input is less than zero otherwise it returns the given input. '''
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1
```

```python
x = np.linspace(-10, 10)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.plot(x, ReLu(x))
plt.axis('tight')
plt.title('Activation Function : RELU')
plt.show()
```

##### Leaky ReLu Activation Function

```python
def LeakyReLu(x):
  ''' It returns a little portion of the input if the input is less than zero otherwise it returns the given input. '''
  x1=[]
  for i in x:
    if i>0 :
      x1.append(i)
    else :
      x1.append(0.1*i)
  return x1
```

```python
x = np.linspace(-10, 10)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.plot(x, LeakyReLu(x))
plt.axis('tight')
plt.title('Activation Function : RELU')
plt.show()
```

##### Softmax Activation Function

```python
def softmax(x):
    ''' Compute softmax values for each sets of scores in x. '''
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
x = np.linspace(-10, 10)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation Function : Softmax')
plt.show()
```

#### backpropagation

$x$ = input(vector of features)

$y$ = target output

$C$ = loss function(cost function)

$L$ = number of layers

$l$ = hidden layer

$W^l$ = weights between layer $l-1$ and $l$, where $w_{jk}^l$ is the weight between the $k$ th node in layer $l-1$ and the $j$ th neuron in layer $l$

$f^l$ = activation functions at layer $l$

$z^l$ = weighted input of each hidden layer

$a^l$ = activation(output) of each hidden layer

$\delta^l$ = gradient of the weighted input of each layer

$g(x) := f^L(W^Lf^{L-1}(W^{L-1} \cdots f^1(W^1x)\cdots))$

$C(y_i, g(x_i))$

$\frac{d C}{d a^L} \odot \frac{d a^L}{d z^L} \cdot \frac{d z^L}{d a^{L-1}} \odot \frac{d a^{L-1}}{d z^{L-1}} \cdot \frac{d z^{L-1}}{d a^{L-2}} \odot \dots \odot \frac{d a^1}{d z^1} \frac{\partial{z^1}}{\partial{x}}$
